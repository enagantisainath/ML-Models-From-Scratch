# Polynomial Regression - Simple Example

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Generate Input and Output data using numpy
X= np.linspace(1,10,20).reshape(-1,1)
Y = X**2+3
```

```python
# Plot
plt.scatter(X,Y)
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```

#### Using Normal Features

```python
model = LinearRegression()
model.fit(X.reshape(-1,1),Y)

plt.scatter(X,Y,color='blue')
plt.plot(X,model.predict(X.reshape(-1,1)),color='red')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```

#### Using Polynomial Features

```python
poly= PolynomialFeatures(degree=4)
model_poly = poly.fit_transform(X.reshape(-1, 1))
```

```python
model = LinearRegression()
model.fit(model_poly,Y)

plt.scatter(X,Y)
plt.plot(X,model.predict(model_poly),color='red')
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```
